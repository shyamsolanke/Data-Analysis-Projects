```python
# Loading data
import pandas as pd
df = pd.read_csv("traffic.csv")
```

```python
# Converting string to datetime
import datetime
df["DateTime"]  = df["DateTime"].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
```

```python
# Checking for Missing values
null_values = df.isnull().sum()
null_values
```

no missing values found

```python
# making datatime column as index
df.index = df["DateTime"] 
```

```python
import holidays
us_holidays = holidays.US(years=[2015, 2016, 2017])
# Extracting Features

df['date'] = df["DateTime"].dt.date
df['hour'] = df['DateTime'].dt.hour
df['weekday'] =  df["DateTime"].dt.strftime("%A")
df['month'] = df['DateTime'].dt.month
df['year'] = df['DateTime'].dt.year
df['isholiday'] = df["date"].isin(us_holidays)
```

```python
# Splitting data by Junctions
df_1 = df[df["Junction"]==1]
df_2 = df[df["Junction"]==2]
df_3 = df[df["Junction"]==3]
df_4 = df[df["Junction"]==4]
```

```python
# Perform the seasonal decomposition
from statsmodels.tsa.seasonal import STL 

# here we kept period as 7 as to check seasonality weekly
advanced_decomposition = STL(df_1.groupby(by="date")["Vehicles"].sum(), period=8).fit()

fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=4, ncols=1, sharex=True)

ax1.plot(advanced_decomposition.observed)
ax1.set_ylabel('Observed')

ax2.plot(advanced_decomposition.trend)
ax2.set_ylabel('Trend')

ax3.plot(advanced_decomposition.seasonal)
ax3.set_ylabel('Seasonal')

ax4.plot(advanced_decomposition.resid)
ax4.set_ylabel('Residuals')

fig.autofmt_xdate()
plt.tight_layout()
```

1. First we are creating a dataset which tells how many cars passed through Junction 1 daily
2. In First SARIMA model we will be only using date to train model
3. In the model after that we will also add other Extracted features to test if that helps in model improving model accuracy 

```python
## create train data
# df_1.reset_index()
df_daily = df_1.resample('D').mean(numeric_only=True)


split_date = '1-Feb-2017'
df_train = df_daily.loc[df_daily.index <= split_date].copy()
df_test = df_daily.loc[df_daily.index > split_date].copy()

df_train.drop(['Junction', 'ID', 'hour', 'month', 'year', 'isholiday'], axis=1, inplace=True)
df_test.drop(['Junction', 'ID', 'hour', 'month', 'year', 'isholiday'], axis = 1, inplace=True)

df_test \
    .rename(columns={'Vehicles': 'TEST SET'}) \
    .join(df_train.rename(columns={'Vehicles': 'TRAINING SET'}),
          how='outer') \
    .plot(figsize=(10, 5), title='Vehicles - Junction 1', style='-', ms=1)
plt.xlabel('Date')
plt.ylabel('Vehicles')
plt.show()
```

1. In the previous PACF graphs we got to know that lag 7, 8, 9 are affecting the time series the most so here in hyper parameter tuning we will tune model from 5 to 11 and see which parameter gives best results
2. First order differencing is making model stable hence trying from d=0 to d=2

```python
## we will now run SARIMAX
'''here we are trying to get best hyperparameters so we use auto_arima, its '''

from pmdarima import auto_arima

model = auto_arima(df_train, 
                   seasonal=True, 
                   m=12,       # 12 months make a season; cycle
                   start_p=0, max_p=11,  # Adjust AR parameter range as we found till lag 8 all can be useful we will use max as 8
                   start_q=0, max_q=2,  # Adjust MA parameter range keeping it low
                   start_d=0, max_d=2,  # Differencing range, usually by second difference the seasonality goes we can also verify this by dickey fuller
                   stepwise=True)
print(model.summary())
```

```python
model.params
```

Best parameters to train on are ARIMA order=(9, 1, 1), seasonal_order=(0, 0, 2, 12)

```python
from sklearn.metrics import mean_squared_error

# Forecast for the length of the test data
forecast = model.predict(n_periods=len(df_test))
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(df_test, forecast))
print(f"RMSE: {rmse}")

mae = np.mean(np.abs(df_test["Vehicles"] - forecast))
print(f"MAE: {mae}")

mape = np.mean(np.abs((df_test["Vehicles"] - forecast) / df_test["Vehicles"])) * 100
print(f"MAPE: {mape}%")
```

```python
# using external variables and keeping best determined hyperparameters
from pmdarima import auto_arima
xternal = df_daily.loc[df_daily.index <= split_date].copy()
xternal = xternal.drop(["Vehicles"], axis=True)
df_train_ex = xternal.loc[xternal.index <= split_date].copy()
model = auto_arima(df_train, 
                   seasonal=True, 
                   m=12, # 12 months make a season; cycle
                   p=9,  # 
                   q=1,  # 
                   d=1,  # 
                   xreg = df_train_ex,
                   stepwise=True)
print(model.summary())
```
